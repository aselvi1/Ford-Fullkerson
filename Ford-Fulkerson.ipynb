```python
# Ford Fullkerson Algoritması
# için Python program uyglaması
from collections import defaultdict

# Bu sınıf yönlü bir grafı temsil
# matris yapısı kullanarak temsil etmektedir.
class Graph:

    def __init__(self, graph):
        self.graph = graph # kalan graf
        self. ROW = len(graph)
        # self.COL = len(gr[0])

    '''Kaynak 's'den, hedef 't' kalıntı grafından bir yol varsa doğruyu döndürür.
     Ayrıca yolu depolamak için parent[]'i doldurur '''

    def BFS(self, s, t, parent):

        #Tüm köşeleri ziyaret edilmemiş olarak işaretle
        visited = [False]*(self.ROW)

        # BFS için yeni kuyruk oluştur
        queue = []

        # Kaynak düğümü ziyaret edilmiş olarak işaretle ve sıraya al
        queue.append(s)
        visited[s] = True

        # Standart BFS döngüsü
        while queue:

            # Bir köşeyi kuyruktan çıkar ve yazdır
            u = queue.pop(0)

            # Sıradan çıkarılan u köşesinin tüm bitişik köşelerini al
            # Eğer bir bitişik köşe ziyaret edilmemişse, o zaman onu işaretle
            # ziyaret edilmişler ve listesine ekle
            for ind, val in enumerate(self.graph[u]):
                if visited[ind] == False and val > 0:
                    # Eğer hedef düğümüne bir bağlantı bulursak,
                    # o zaman artık BFS'nin bir anlamı kalmaz
                    # Sadece ebeveynini ayarla ve true döndür
                    queue.append(ind)
                    visited[ind] = True
                    parent[ind] = u
                    if ind == t:
                        return True

        # Kaynaktan başlayarak BFS'de hedefe ulaşamadık
        # bu yüzden false döndür

        return False


    # Verilen grafta s'den t'ye maksimum akışı döndürür
    def FordFulkerson(self, source, sink):

        # Bu dizi BFS tarafından doldurulur ve paterni depolamak için kullanılır
        parent = [-1]*(self.ROW)

        max_flow = 0 # Başlangıçta akış yok

        # Kaynak ile hedef arasında bir yol varken akışı artır
        while self.BFS(source, sink, parent) :

            # BFS tarafından doldurulan
            # yol boyunca kenarların minimum kalan kapasitesini bulur.
            # Ya da bulunan yol boyunca maksimum akışı bulur diyebiliriz.
            path_flow = float("Inf")
            s = sink
            while(s != source):
                path_flow = min (path_flow, self.graph[parent[s]][s])
                s = parent[s]

            # Yol akışını genel akışa ekler
            max_flow += path_flow

            # kenarların ve ters kenarların kalan kapasitelerini güncelle
            # yol boyunca
            v = sink
            while(v != source):
                u = parent[v]
                self.graph[u][v] -= path_flow
                self.graph[v][u] += path_flow
                v = parent[v]

        return max_flow


# Yukarıdaki diyagramda verilen grafı oluştur
"""
graph = [[0, 16, 13, 0, 0, 0],
        [0, 0, 10, 12, 0, 0],
        [0, 4, 0, 0, 14, 0],
        [0, 0, 9, 0, 0, 20],
        [0, 0, 0, 7, 0, 4],
        [0, 0, 0, 0, 0, 0]]
"""
graph = [[0, 14, 12, 0, 0, 0],
        [0, 0, 0, 10, 0, 0],
        [0, 6, 0, 0, 11, 0],
        [0, 0, 7, 0, 0, 18],
        [0, 0, 0, 5, 0, 6],
        [0, 0, 0, 0, 0, 0]]
g = Graph(graph)

source = 0; sink = 5

print ("Maksimum akış kapasitesi %d " % g.FordFulkerson(source, sink))
```
